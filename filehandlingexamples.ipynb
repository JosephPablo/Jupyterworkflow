```python
#using a context manager to close files

with open('joseph.txt','r') as f:
    pass
print(f.read())# This statement will throw an error because the operation is performed on a closed file 
```

```python
#To read the contents of the file
with open('joseph.txt','r') as f:
    f_contents = f.read()
    print(f_contents)
    
```

# working with CSV files

source:https://www.youtube.com/watch?v=q5uM4VKywbA

```python
#importing a csv file into notebook
import csv

with open('sde_20161201.csv','r')as csv_file:
    csv_reader = csv.reader(csv_file)
    
    for line in csv_reader:
        print(line)
        
    
```
