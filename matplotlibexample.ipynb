# MATPLOTLIB EXAMPLE

```python
#import packages
import matplotlib.pyplot as plt
import numpy as np

```

```python
#make calculations
t = np.arange(0.0, 2.0, 0.01)# Start_value=0.0,stop_value=2.0,step_value=0.01
s = 1 + np.sin(2*np.pi*t)#formula for sine wave
plt.plot(t, s) #plot(independent_variable,dependent_variable)

```

```python
#PLOTTIING THE COMPLETE GRAPH
plt.xlabel('time (s)')
plt.ylabel('voltage (mV)')
plt.title('Plotting the sine wave')
plt.grid(True)
plt.savefig("sine.png")
plt.show()
```
